```python
class SNumPy():
    """
    SNumPy Class Definition
    
    """
    
    # a function used troughout the class to check if input arrays are vectors i.e., if every element of an array is of type integer
    def vector_check(self, array):
       
        # list comprehension iterating though evey element of an array to see if all elements are of type interger
        return(all(isinstance(i, int) for i in array) == True)
        
     # a function used troughout the class to check if input arrays are matrices i.e., if every element of an array is of type integer    
    def matrix_check(self, array):
        
        # list comprehension iterating though evey element of an array to see if all elements are of type list
        if all(isinstance(i, list) for i in array) == True:
            
            # nested list comprehension iterating though every nested list of the input array and every element of the nested lists to detimne that the nested lists are of equal lenght, and only contains integers as elements
            return(all([all([isinstance(i, int) for i in l]) for l in array]) and all(len(j) == len(array[0]) for j in array) == True)

        else:
            return False

        
    # Exercise 1:
    
    def ones(self, dim):
        """
        Takes a dimension parameter that can be either an integer or a tuple of rows and columns (integers).
        Return an array/list of the lenght of the integer parameter, or a matrix of the input dimensions, only containing ones.
        
        Args:
        dim:  an integer or tuple of integers
        
        Returns:
        Return an array/list of the lenght of the interger parameter, or a matrix of the input dimensions, only containing ones. 
        
        """
        
        # if block to check if input type is interger
        if isinstance(dim, int):
            
            # creating an array of ones by multiplying [1] by the input interger value
            array = [1] * dim
            
            return array
        
        # if block to check if input is tuple
        if isinstance(dim, tuple):
            
            # specifying that dimensions consist of number of rows and number of columns
            nr_rows, nr_cols = dim

            # creating a single row of ones by multiplying number of columns with [1]
            row = nr_cols * [1]

            # creating an empty list
            array = []
            
            # iterating though the range of rows and adds the single row to the empty list as many times as number of rows in dimensions
            for r in range(nr_rows):
                array.append(row)
                
            return array              
        
        # else block that will print an errormessage if the input is not evaluated to be of type interger or tuple
        else:
            print("Input for dim not valid. Please enter only integers or a tuple of integers as an input argument.")
            return
    
    
    # Exercise 2:
    
    def zeros(self, dim):
        """
        Takes a dimension parameter that can be either an integer or a tuple of rows and columns (integers).
        Return an array/list of the lenght of the integer parameter, or a matrix of the input dimensions, only containing zeros.
        
        Args:
        dim:  an integer or tuple of integers
        
        Returns:
        Return an array/list of the lenght of the interger parameter, or a matrix of the input dimensions, only containing zeros. 
        
        """

        # if block to check if input type is interger                
        if isinstance(dim, int):
            
            #creating an array of zeros by multiplying [0] by the input interger value
            array = [0] * dim
            
            return array
        
        # if block to chech if input type is tuple
        if isinstance(dim, tuple):

            # specifying that dimensions consist of number of rows and number of columns
            nr_rows, nr_cols = dim

            # creating a single row of zeros by multiplying number of columns with [0]
            row = nr_cols * [0]

            # creating an empty list
            array = []
            
            # iterating though the range of rows to add the single row to the empty list as many times as number of rows in dimensions
            for r in range(nr_rows):
                array.append(row)
                
            return array
        
        # else block that will print an errormessage if the input is not evaluated to be of type interger or tuple
        else:
            print("Input for dim not valid. Please enter only integers or a tuple of integers as an input argument.")
            return
    
    
    # Exercise 3: 
    
    def reshape(self, array, dimension):
        """
        Converts the array into the dimensions specified in given tuple of row, column.
        
        Args:
        array:  an array in either the shape of a vector or a matrix 
        
        Returns:
        Returns a reshaped vector/matrix in the dimensions specified by the tuple (row, column)
               
        """
        
        # if the array is a matrix, convert it to a one dimentional list
        if self.matrix_check(array) == True:
            array = [item for sublist in array for item in sublist]

        # creating an empty list to append every list of rows to, to get the final matrix
        matrix = []

        # specifying a counter for the interation through rows
        k = 0

        # specifying that the input tuple dimensions are rows and collumns
        rows, cols = dimension

        # if block to check if input array is a valid vector
        if self.vector_check(array) == False:
            
            # if not, print an error
            print("Input for array not valid. Please enter only integers in the array as an input argument.") 
            return
        
        # if rows * columns equals the lenght of the input array, it is appropriate for reshaping and we enter the nested loop
        if rows*cols == len(array):
            
            # the outler loop determines the lenght of the rows, based on the input rows dimension
            for element in range(rows):

                # creating a list of rows from where the inner loop are to append every row element of the matrix 
                list_rows = []

                # the inner loop iterates over the range of the input column dimension and appends every element of the list until the range of the input column is reached
                for element in range(cols):
                    list_rows.append(array[k])

                    # using the addition assignment operator the counter is keeping track of the iterations
                    k += 1

                # when the nested loops are finished iterating, every list/row containing elements from the array is appended to the empty matrix list, and the final matrix is returned with the input dimensions
                matrix.append(list_rows)
            return matrix

        # if rows * columns are not equal the lenght of the input array, an error message will print through the else statement
        else: 
            print("Please enter dimensions appropriate for reshaping.") 
            return
    
    
    # Exercise 4:
    
    def shape(self, array):
        """
        Finds the number of rows and the number of columns from the given array.
        
        Args:
        array:  an array in either the shape of a vector or a matrix 
        
        Returns:
        Returns a tuple with the matrix/vectorâ€™s dimension e.g. (# rows, # columns). 
               
        """
        
        # if block to check if input array is a valid vector
        if self.vector_check(array) == True:

            # the row dimension is defaulted to one, if the input is a vector
            rows = 1

            # the column dimension are simply the lenght of the input array
            cols = len(array)

            # returns dimensions
            return (rows, cols)

        # elif block to check if input array is a valid matrix
        elif self.matrix_check(array) == True:

            # the rows are determined as the lenght of the array, if the input is a matrix
            rows = len(array)

            # the column dimensions are determined by referencing the lenght of the first index element of the array (the first nested list)
            cols = len(array[0])
            
            # returns dimensions
            return (rows, cols)

        # if none of the above type conditions are met, an error message will print through the else statement and guide the user to approriate input.
        else: 
            print("Input for array not valid. Please make sure array only contains integers or a list of equally long nested lists of integers.")
            return
    
    
    # Exercise 5

    def append(self, array1, array2):
        """
        Takes two input vectors/matrices and append them with each other. 
        Obs: You can't append a vector to a matrix and vice versa.
        
        Args:
        array1:  an array in either the shape of a vector or a matrix 
        array2:  an array in either the shape of a vector or a matrix 
        
        Returns:
        Returns a new vector/matrix that is the combination of the two input vectors/matrices.       
        
        """
        
        # using the shape function from above to get the dimensions (row, column) of the input arrays
        shape_array1 = self.shape(array1)
        shape_array2 = self.shape(array2)
        
        # if block to check if input arrays are BOTH valid vectors
        if (self.vector_check(array1) and self.vector_check(array2) == True):
            
            # if valid vectors, perform addition
            return (array1 + array2)
        
        # if block to check if input arrays are BOTH valid matrices
        if (self.matrix_check(array1) == True and self.matrix_check(array2) == True):
            
            # nested if block to check if input arrays have equal column dimensions
            if shape_array1[1] == shape_array2[1]:
                
                # if valid matrices of equal column dimensions, perform addition
                return (array1 + array2)
            
            # if columns of arrays are not of equal lenght, an error message will print through the nested else statement
            else:
                print("Input for array1 and array2 not valid. Please make sure arrays have the same number of columns.")
                return
        
        # if none of the above type conditions are met, an error message will print through the else statement and guide the user to approriate input.
        else: 
            print("Input for array1 and array2 not valid. Please make sure arrays only contains integers or a list of equally long nested lists of integers. Also, not possible to add a vector to a matrix and vice versa.")
            return
    
    
    # Exercise 6:
    
    def get(self, array, row_col):
        """
        A function that returns the value indicated by the coordinate point (row, column) of the given array. 
        
        Args:
        array:  an array in either the shape of a vector or a matrix 
        row_col: a tuple containing the dimensions indicating which index to return

        Returns:
        Returns the value specified by the coordinate point (row, column)
        
        """
        # if block to check if input array is a valid vector
        if self.vector_check(array) == True:
            
            # index 1 is the column of the tuple row_col and index 0 is the row of the tuple
            # returns the value of the vector's specific column index position given in parameter row_col[1] 
            return array[row_col[1]]
        
        # elif block to check if input array is a valid matrix
        elif self.matrix_check(array) == True:
            
            # returns the value of the matrix's specific row and column index positions given in parameter row_col[0] and row_col[1]
            return array[row_col[0]][row_col[1]]
        
        # if none of the above type conditions are met, an error message will print through the else statement and guide the user to approriate input
        else:
            print("Input for array not valid. Please make sure array only contains integers or a list of equally long nested lists of integers.")
            return

        
    # Exercise 7:
    
    def add(self, array1, array2):
        """
        Addition is performed on vectors and matrices.
        It takes each value from the vectors or matrices and adds it to each other.
        
        Args:
        array1:  an array in either the shape of a vector or a matrix 
        array2:  an array in either the shape of a vector or a matrix 
        
        Returns:
        Returns the result of either two added vectors or matrices. 
        
        """

        # using the shape function from above to get the dimensions (row, column) of the input arrays
        shape_array1 = self.shape(array1)
        shape_array2 = self.shape(array2)
        
        # if block to check that the dimensions of the input arrays are equal 
        if shape_array1 == shape_array2:
        
            # nested if block to check if the given arrays are vectors 
            if self.vector_check(array1) == True and self.vector_check(array2) == True:

                # adding two vectors using list comprehension; in a row-vector you only need to add the values of each column with each other
                return([array1[x] + array2[x] for x in range(shape_array1[1])])
            
            # nested elif block to check if the given arrays are matrices  
            elif self.matrix_check(array1) == True and self.matrix_check(array2) == True:

                # adding two matrices in a nested loop using list comprehension; i and j takes each row and column and add the value with each other
                return([[array1[i][j] + array2[i][j] for j in range(shape_array1[1])] for i in range(shape_array1[0])])
            
            # if none of the above type conditions are met, an error message will print through the nested else statement and guide the user to approriate input
            else: 
                print("Input for array1 and array2 not valid. Please make sure arrays only contain integers or a list of equally long nested lists of integers (Obs. dimensions of the input arrays must be equal!)")
                return
        
        # if the dimensions (row, column) of the input arrays are not equal, an error message will print throught the outer else statement highlighting the mathematical conditions of the function
        else: 
            print("Not possible to add a vector to a matrix and vice versa. Please only input two vectors or two matrices of same dimensions.")
            return
        
        
    # Exercise 8:
    
    def subtract(self, array1, array2):
        """
        Subtraction is performed on vectors and matrices. 
        It takes each value from the vectors or matrices and subtract from each other.
        
        Args:
        array1:  an array in either the shape of a vector or a matrix 
        array2:  an array in either the shape of a vector or a matrix 
        
        Returns:
        Returns the result of either two subtracted vectors or matrices. 
        
        """   
    
        # using the shape function from above to get the dimensions (row, column) of the input arrays
        shape_array1 = self.shape(array1)
        shape_array2 = self.shape(array2)
        
        # if block to check that the dimensions of the input arrays are equal 
        if shape_array1 == shape_array2:
        
            # nested if block to check if the given arrays are vectors 
            if self.vector_check(array1) == True and self.vector_check(array2) == True:

                # subtracting two vectors using list comprehension; in a row-vector you only need to subtract the values of each column with each other
                return([array1[x] - array2[x] for x in range(shape_array1[1])])
            
            # nested elif block to check if the given arrays are matrices 
            elif self.matrix_check(array1) == True and self.matrix_check(array2) == True:

                # subtracting two matrices in a nested loop using list comprehension; i and j takes each row and column and subtract the value with each other
                return([[array1[i][j] - array2[i][j] for j in range(shape_array1[1])] for i in range(shape_array1[0])])
            
            # if none of the above type conditions are met, an error message will print through the nested else statement and guide the user to approriate input
            else: 
                print("Input for array1 and array2 not valid. Please make sure arrays only contain integers or a list of equally long nested lists of integers (Obs. dimensions of the input arrays must be equal!)")
                return
        
        # if the dimensions (row, column) of the input arrays are not equal, an error message will print throught the outer else statement highlighting the mathematical conditions of the function        
        else: 
            print("Not possible to subtract a vector from a matrix and vice versa. Please only input two vectors or two matrices of same dimensions.")
            return
            

    # Exercise 9:
    
    def dotproduct(self, array1, array2):
        """
        Calculates the dot product between the two given arrays, which could be either vectors or matrices, 
        and then returns a suitable value.
        
        Args:
        array1:  an array in either the shape of a vector or a matrix 
        array2:  an array in either the shape of a vector or a matrix 
        
        Returns:
        Returns the result of the dot product between two arrays (vectors or/and matrices) and returns an appropriate value.
        """
        
        # using the shape function from above to get the dimensions (row, column) of the input arrays
        row1, col1 = self.shape(array1)
        row2, col2 = self.shape(array2)

        # a new empty list to be returned containing the results after doing dot product on the input arrays
        matrix = []

        # if block to check if input arrays are BOTH valid vectors AND that the number of columns of the two are equal
        if self.vector_check(array1) == True and self.vector_check(array2) == True and col1 == col2:

            sum_vector = 0

            # iiterate through the columns of the vector (array1)
            for i in range(col1):
                
                #execute the dot-product on vector; which takes the sum of the products of array1 and array2 and stores the result in the sum_vector variable
                sum_vector += array1[i] * array2[i]

            # returning the dot product of two vectors 
            return sum_vector

        # elif block to check if array1 is a valid vector or matrix AND if array2 is a valid vector or matrix; dot product is allowed on vector and matrix
        # an additional AND checks that the number of columns in the first array is the same as the number of rows in the second array
        elif (self.vector_check(array1) == True or self.matrix_check(array1) == True) and (self.vector_check(array2) == True or self.matrix_check(array2) == True) and col1 == row2:
            
            # we convert all vectors to nested lists in order to execute dot product on a vector with a matrix
            
            # if block to check if array1 is a valid vector
            if self.vector_check(array1):
                
                # convert vector to a nested list
                array1 = [array1]
            
            # if block to check if array2 is a valid vector
            if self.vector_check(array2):
                
                # convert vector to a nested list
                array1 = [array2]
                
            # iterate through the rows of array1
            for i in range(row1):
                
                # an empty list generated in each outer iteration used by the inner iteration creating the rows of the matrix
                l = []

                # iterate through the columns of array2
                for j in range(col2):
                    
                    sum_matrix = 0

                    # keeping the row and the column constant from array1 and array2
                    # iterate through the rows in the nested list of array2 
                    for x in range(row2):

                        # the dot product is the sum of the product of each value in each given index position from array1 and array2
                        sum_matrix += array1[i][x] * array2[x][j]

                    # adding the output from the dot product to the local list variable
                    l.append(sum_matrix)

                # adding the list to the matrix which generates a new row in each iteration
                matrix.append(l)
            
            return matrix
        
        # if none of the above type conditions are met, an error message will print through the nested else statement and guide the user to approriate input
        else: 
            print("Input for array1 and array2 not valid. Please make sure arrays only contain integers or a list of equally long nested lists of integers (Obs. dimensions of the input arrays must be equal!)")
            return
```
